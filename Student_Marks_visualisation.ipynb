```python
# Generate data
import pandas as pd
import numpy as np

np.random.seed(1)  # for reproducibility
students = [f"Student_{i+1}" for i in range(20)]
marks = {
    "Math": np.random.randint(40, 100, 20),
    "Science": np.random.randint(40, 100, 20),
    "English": np.random.randint(40, 100, 20),
    "History": np.random.randint(40, 100, 20),
}
df = pd.DataFrame(marks, index=students)
df.head()

```

```python
# Bar chart
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.bar(df.index, df["Math"], color='skyblue', edgecolor='black')
plt.xticks(rotation=90)
plt.title("Math Scores of Students")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Line Chart
student = "Student_5"
plt.plot(df.columns, df.loc[student], marker='o', linestyle='--', color='red', label=student)
plt.title(f"Scores of {student}")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Histogram
plt.hist(df["Science"], bins=8, color="green", edgecolor="black")
plt.title("Distribution of Science Scores")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.grid(True)
plt.show()
```

```python
#Boxplot
plt.boxplot([df["Math"], df["Science"], df["English"], df["History"]],
            tick_labels=["Math", "Science", "English", "History"])
plt.title("Boxplot of Marks by Subject")
plt.ylabel("Scores")
plt.grid(True)
plt.show()
```
